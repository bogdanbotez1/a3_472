```python
def legal_move_eval(taken_Token, past_Tokens):

    if((taken_Token % past_Tokens == 0) or (past_Tokens % taken_Token == 0)):
        
        return True
    else:
        return False
```

```python
def legal_move(taken_Token, past_Tokens, tokenAvailable):
    if(len(past_Tokens) == 0):
        
        if(len(tokenAvailable)//2 >= taken_Token):
            return True
        else:
            return False

    if((taken_Token % past_Tokens[-1] == 0) or (past_Tokens[-1] % taken_Token == 0)):
        
        return True
    else:
        return False

        
print(legal_move(4,[1,2,3],[4,5,6,7]))

```

```python

general_depth =0 
nodes_Visited =1 
nodes_Evaluated =0 

def PNT_Player(tokens, taken_tokens , list_of_taken_tokens , depth, max_depth):
    
    if(max_depth == 0):
        max_depth = 99
    global general_depth
    general_depth =0 
    
    global nodes_Visited
    nodes_Visited =1 
    
    global nodes_Evaluated
    nodes_Evaluated =0 
    
    list_tokens_available = []
    for token in range(tokens):
        list_tokens_available.append(token+1)
         
    for token in list_of_taken_tokens:
        for tokenX in list_tokens_available:
            if(token == tokenX):
                list_tokens_available.remove(token)
                
#     print(list_tokens_available)
#     print(list_of_taken_tokens)

    
    
    if(taken_tokens % 2 == 0): # MAX TURN
        
        x = max_alpha_beta(list_of_taken_tokens,-99,99,list_tokens_available,depth,max_depth)
        nodes_Evaluated = nodes_Visited- nodes_Evaluated

        print("")
        print("(Value, Move): " + "")
        print(x)

        print("nodes_Visited: " + str(nodes_Visited))

        print("nodes_Evaluated: " + str(nodes_Evaluated))
        print("MAX DEPTH: " + str(general_depth))
        if(nodes_Visited - nodes_Evaluated) == 0:
            print("avgBranchFactor: division by zero")
        else:
            avgBranchFactor = (nodes_Visited - 1) / (nodes_Visited - nodes_Evaluated)
            print("avgBranchFactor: " + str(round(avgBranchFactor,1)))
    else:                      # MIN TURN
        x = min_alpha_beta(list_of_taken_tokens,-99,99,list_tokens_available,depth,max_depth)
        
        nodes_Evaluated = nodes_Visited- nodes_Evaluated
        print("")
        print("(Value, Move): " + "")
        print(x)

        print("nodes_Visited: " + str(nodes_Visited))

        print("nodes_Evaluated: " + str(nodes_Evaluated))
        print("MAX DEPTH: " + str(general_depth))
        
        if(nodes_Visited - nodes_Evaluated) == 0:
            print("avgBranchFactor: 1") #division by zero so ==1
        else:
            avgBranchFactor = (nodes_Visited - 1) / (nodes_Visited - nodes_Evaluated)
            print("avgBranchFactor: " + str(round(avgBranchFactor,1)))

    
    
PNT_Player(3,0,[],0,0)    
    
PNT_Player(7,1,[1],0,2)


PNT_Player(10,3,[4,2,6],0,4)



PNT_Player(7,2,[1,2],0,0)



PNT_Player(10,0,[],0,0)


PNT_Player(4,1,[1],0,0)

PNT_Player(8,2,[3,6],0,0)

PNT_Player(11,3,[4,8,2],0,0)

PNT_Player(20,0,[],0,0)

PNT_Player(20,0,[],0,5)
```
